```python
import cv2
import numpy as np

img = cv2.imread("test images/melanoma_me.jpg") #load image
img = cv2.resize(img, (300, 300)) #resize image to 300 x 300
grayImg = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY) #grayscale version
```

```python
_ , thresh = cv2.threshold(grayImg, np.mean(grayImg), 255, cv2.THRESH_BINARY_INV)  # '_' symbol is used as an ignore symbol
#thresh = cv2.adaptiveThreshold(grayImg, 255, cv2.ADAPTIVE_THRESH_MEAN_C, cv2.THRESH_BINARY_INV, 81, 2)
# get contours
contours, hierarchy = cv2.findContours(thresh, cv2.RETR_LIST, cv2.CHAIN_APPROX_SIMPLE)

print(len(contours))

# lets get the bigger area 
cnt = sorted(contours, key=cv2.contourArea)[-1]

#print dimensions and color channels
print(img.shape)
```

```python
mask = np.zeros((300, 300), dtype = "uint8")
#imageMask1 = cv2.drawContours(mask, [cnt] , -1 , (0, 0, 255), -1)
imageMask2 = cv2.drawContours(mask, [cnt] , -1 , (255, 255, 255), -1)

finalImage = cv2.bitwise_and(img, img, mask=imageMask2) 
```

```python
cv2.imshow("binary mask", imageMask2)
cv2.imshow("masked image", finalImage)
cv2.imshow("Original", img)
cv2.waitKey(0) #waits forever until user presses a key
cv2.destroyAllWindows() #destroys windows when the wait is done
```
